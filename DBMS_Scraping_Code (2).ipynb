```python
# This code to get quarterly information was run 4 times to get all the data
# This is because continuous pinging led to '403 Forbidden Error'

Quarterly_Data1 = pd.DataFrame()
j=0
for i in url:
    driver = webdriver.Chrome()
    
    driver.get(i)
    driver.implicitly_wait(15)
    soup = BeautifulSoup(driver.page_source, 'lxml')
    tables = soup.find_all('table')
    
    # Step 3: Read tables with Pandas read_html()
    dfs = pd.read_html(str(tables))
    
    list1 = dfs[0]
    df1 = pd.DataFrame(list1)
    
    new_header = df1.iloc[0] #grab the first row for the header
    df1 = df1[1:] #take the data less the header row
    df1.columns = new_header #set the header row as the df header
    
    df1['Ticker'] = tckr[j]
    Quarterly_Data1 = Quarterly_Data1.append(df1, ignore_index=True)
    j=j+1

Quarterly_Data1.head()
```

### 2. Getting sector, industry, number of follower and number of analysts information

```python
Company_Df = pd.DataFrame()
for i in url:
    driver = webdriver.Chrome()
    driver.get(i)
#driver.get('https://www.estimize.com/amat/fq4-2022?metric_name=eps&chart=historical')
    driver.implicitly_wait(15)
    soup = BeautifulSoup(driver.page_source, 'lxml')
    
    TICKER = soup.find("h1", class_="release-header-information-title").get_text(strip=True)
    NAME = soup.find("p", class_="release-header-information-description").get_text(strip=True)

    SECTOR_INDUSTRY = soup.find("p", class_="release-header-information-breadcrumb").get_text(strip=True)
    SECTOR = SECTOR_INDUSTRY.split('›')[0]
    INDUSTRY = SECTOR_INDUSTRY.split('›')[1]

    NUMBER_FOLLOWERS = soup.find("div", class_="estimize-row summary-sub-title").get_text()
    NUMBERS_ANALYSTS = soup.find("a", class_="estimize-row summary-sub-title").get_text()

    Company_Temp = {'Ticker': TICKER,
                    'Name': NAME,
                    'Sector': SECTOR,
                    'Industry': INDUSTRY,
                    'Num_Followers': NUMBER_FOLLOWERS,
                    'Num_Analysts': NUMBERS_ANALYSTS}

    Company_Df = Company_Df.append(Company_Temp, ignore_index=True)

driver.close()
Company_Df
```

```python
driver = webdriver.Chrome()
driver.get('https://www.estimize.com/amat/fq4-2022?metric_name=eps&chart=historical')
driver.implicitly_wait(25)
soup = BeautifulSoup(driver.page_source, 'lxml')

tables = soup.find_all('table')
dfs = pd.read_html(str(tables))

list1 = dfs[4]
df1 = pd.DataFrame(list1)

df1 = df1[df1['Rank'].notna()] #take the data less the header row
df1['Analyst'] = df1['Analyst'].apply(lambda x: x[:len(x)//2] if len(x)%2 == 0 else x[len(x)//2:])
df1['Analyst']=df1['Analyst'].str.lower()

list2 = df1['Analyst'].values.tolist()
list2 = list2[0:3]

url=[]
for n in list2:
    url1 = ['https://www.estimize.com/users/' + n]
    url.append(url1)
           
driver.close()
```

```python
Analyst_Df = pd.DataFrame()
for i in url:
    driver = webdriver.Chrome()
    driver.get(i)
    driver.implicitly_wait(15)
    soup = BeautifulSoup(driver.page_source, 'lxml')
    
    ANALYST_NAME=soup.find("h1", class_="profile-display-name").get_text(strip=True)
    ROLES=soup.find("ul", class_="profile-bio-categorizations").get_text(strip=True)
    JOINING_DATE=soup.find("div", class_="profile-activity-stats").get_text(strip=True)
    JOINING_DATE = JOINING_DATE[13:21]
    
    CONF_SCORE=soup.find("div", class_="value").get_text(strip=True)
    
    TEMP1=soup.find("div", class_="profile-tab profile-tab-selectable profile-tab-accuracy is-selected").get_text()
    ACCURACY_PERC=TEMP1.split()[9]
    ERROR_RATE=TEMP1.split()[3]
    
    TEMP2 =soup.find("div", class_="profile-tab profile-tab-selectable not-selected profile-tab-points").get_text()
    POINTS = TEMP2.split()[2]
    POINTS_ESTIMATE = TEMP2.split()[7]
    
    TEMP3 =soup.find("div", class_="profile-tab profile-tab-unselectable profile-tab-last not-selected profile-tab-stocks").get_text()
    STOCKS = TEMP3.split()[2]
    PENDING = TEMP3.split()[8]
    
    Analyst_Temp = {'Name': ANALYST_NAME,
                    'Role': ROLES,
                    'Joining Date': JOINING_DATE,
                    'Confidence Score': CONF_SCORE,
                    'Accuracy Percentile': ACCURACY_PERC,
                    'Error rate': ERROR_RATE,
                    'Points': POINTS,
                    'Ponts Estimate': POINTS_ESTIMATE,
                    'Stocks': STOCKS,
                    'Pending': PENDING}

    Analyst_Df = Analyst_Df.append(Analyst_Temp, ignore_index=True)
    
driver.close()
Analyst_Df
    
```
